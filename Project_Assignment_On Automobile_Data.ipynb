### Import Libraries.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.simplefilter('ignore')
```

```python
data = pd.read_csv('automobile/Automobile_data.csv')
data.head()
```

```python
data.describe()
```

### Data Cleaning.

```python
data = data.drop('index', axis=1)
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data.price = data.price.fillna(data.price.mean())
```

## 1.

a)
### Most expensive car company name.

```python
data.company[data.price.idxmax()]
```

b)
### Details of all Toyota cars.

```python
data[data.company=='toyota']
```

c)
### Total cars per company

```python
data.company.value_counts()
```

## 2.

### The car of each company which costs highest among other cars, and the average mileage providing car company name.

```python
data.groupby('company').max()
```

## 3.
### All companies producing the cars with 'num-of-cylinders'is equal to'six'

```python
print(data['company'][data['num-of-cylinders']=='six'].values)
```

## 4. 

### All companies details which are producing the cars with 'average-mileage' greater than or equal to 35 and number of cylinders equal to four.

```python
data[(data['average-mileage']>=35) & (data['num-of-cylinders']=='four')]
```
